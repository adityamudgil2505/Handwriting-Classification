```python
def printPokemon(numbers):
    pokemon_map = {
    1: "Bulbasaur",
    2: "Pikachu",
    3: "Charmender"
    }
    for i in numbers:
        if pokemon_map.get(i) is None:
            print("Not Present")
        else:
            print(pokemon_map[i])
    return
num = [int(x) for x in input().split()]
printPokemon(num)
```

```python
import matplotlib.pyplot as plt
x = np.array([1,2,3,4])
y = x**2 + 3
plt.plot(x, y)
plt.show()
```

```python
# plt.style.available
plt.style.use('seaborn')
plt.plot(x, y, label="Stock Price")
plt.plot(x, y**1.5, label="Stock Price 2")
plt.title("My first graph")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use('dark_background')
plt.scatter(x, y, label="Stock Price")
plt.scatter(x, y**1.5, label="Stock Price 2")
plt.title("My first scatter graph")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img1.png')
print(img.shape)
plt.imshow(img)
plt.show()
# bgr format of opencv when we input image it conver red to blue and vise versa
# rbg which we want to display
# converting bgr to rbg
rbg_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis("off") # to remove extra labeling
plt.imshow(rbg_img)
plt.show()

# cropping image
crop_img = rbg_img[250:850, 150: 950,:]
plt.imshow(crop_img)
plt.show()
```

```python
totalPixels = img.shape[0] * img.shape[1]
print(totalPixels)

# Resize image
resized_img = cv2.resize(rbg_img, (100, 100))
plt.imshow(resized_img)
plt.show()
```
