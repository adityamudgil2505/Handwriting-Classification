```python
# calculating distance function

def dist(a, b):
    return np.sum((a-b)**2)**.5
```

```python
# K-nearest neighbour Algorithm

def KNN(X, Y, test_point, k=5):
    m = X.shape[0]
    val=[]
    for i in range(m):
        d = dist(test_point, X[i])
        val.append((d, Y[i])) # this will store distance and the y label 
    val=sorted(val)
    val=np.array(val[:k])
    temp=np.unique(val[:,1],return_counts=True)
    idx=np.argmax(temp[1])
    res=temp[0][idx]
    return int(res)
```

```python
# KNN on handwritten Digit Classification (MNIST)
import pandas as pd;
df=pd.read_csv('train.csv')
# print(df.head())
```

```python
# We can't do slicing on dataframe therefore converting dataframe to np array
data=df.values
Y=data[:,0]
X=data[:,1:]
```

```python
# testing using KNN algorithm
i=19
test_sample=X[i]
pred=KNN(X,Y,test_sample)
actual=Y[i]
plt.imshow(X[i].reshape([28,28]),cmap='gray')
plt.show()
print("pred = ",pred)
print("actual = ",actual)
```
